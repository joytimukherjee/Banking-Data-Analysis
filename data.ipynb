NUMERICAL ANALYSIS

```python


# Numerical analysis and exploration
numerical_cols = ['Fee Structure','Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Univariate analysis and visualization
py.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    py.subplot(4, 3, i + 1)
    sns.histplot(df[col], kde=True)
    py.title(col)
py.tight_layout()
py.show()
```

```python
# Select numerical columns for correlation analysis
numerical_cols = ['Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
                  'Foreign Currency Account', 'Business Lending', 'Properties Owned']

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Create a heatmap of the correlation matrix
py.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
py.title('Correlation Matrix of Numerical Features')
py.show()
```

1.High Positive Correlation Among Account Types
Bank Deposits show strong positive correlation with:
Checking Accounts
Saving Accounts
Foreign Currency Accounts
Suggests customers with high balance in one account type usually hold significant funds in others too.

2.Income, Age, and Accumulation
Age and Estimated Income show moderate correlation with:
Superannuation
Savings
Checking balances
Reflects typical financial lifecycle: older and higher-income individuals accumulate more assets and loans.

3.Weak Correlation with Property Ownership
Property variables show low correlation with banking metrics.
Indicates ownership depends on external factors (e.g., market, inheritance) not captured here.
Business vs. Personal Banking Behavior
Moderate correlation between Business Lending and Bank Loans.
Business lending is mostly independent of deposits/property metrics — hinting at a distinct customer segment.
